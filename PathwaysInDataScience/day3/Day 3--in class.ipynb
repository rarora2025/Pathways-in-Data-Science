```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
samples = 50
x_og = np.random.normal(size=samples)
y_og = 2*x_og + np.random.normal(scale = 0.1,size=samples)

for outlier in [[0,8],[10,10],[10,10*2+np.random.normal(scale=0.1)]]:
    x = np.append(x_og,outlier[0])
    y = np.append(y_og,outlier[1])

    plt.plot(x, y, 'o', label='original data')
    plt.title('Original data')
    plt.legend()
    plt.show()

    res = stats.linregress(x, y)
    plt.plot(x, y, 'o', label='original data')
    plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line')
    plt.title(f"Fitting including outlier \n m = {res.slope:.6f} \n R-squared: {res.rvalue**2:.6f}")
    plt.legend()
    plt.show()

    res = stats.linregress(x[:-1], y[:-1])
    plt.plot(x, y, 'o', label='original data')
    plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line')
    plt.title(f"Fitting EXCLUDING outlier \n m = {res.slope:.6f} \n R-squared: {res.rvalue**2:.6f}")
    plt.legend()
    plt.show()
```
